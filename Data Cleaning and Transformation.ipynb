```python
import pandas as pd
import numpy as np


exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas', "Elias", 'Myriam'],

'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan , 8, 19, 20, 16],

'attempts': [1, 3, 2, 3, 2, 3, 1, 1, np.nan ,1, 3, 2],

'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no',np.nan , 'yes', 'yes', 'yes']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l']

df = pd.DataFrame(exam_data)
df1 = pd.DataFrame(labels, columns=['labels'])
df1
df=df.join(df1)
df.set_index("labels",inplace=True)
df

```

```python
df.isnull().sum()
```

```python
df['score'].fillna(df['score'].mean(),inplace=True)
df
```

```python

df['attempts'].fillna(df['attempts'].mean(),inplace=True)
df

```

```python
print(df['qualify'].value_counts())
```

```python
df['qualify'].fillna('yes',inplace=True)
df
```

```python
one_hot=pd.get_dummies(df['qualify'])
df.drop('qualify',axis=1,inplace=True)
df=df.join(one_hot)
df
```

```python
list = ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no',np.nan , 'yes', 'yes', 'yes']
serie = pd.Series(list)
serie
```

```python
df.drop('no',axis=1,inplace=True)
df.drop('yes',axis=1,inplace=True)
df=df.join(Serie)
df
```
